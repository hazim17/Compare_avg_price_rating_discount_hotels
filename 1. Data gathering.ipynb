```python
# Hotel rating
ratings = soup.find_all("div", class_="rat")

rating_hotel_num = []
rating_hotel_text = []
rating_hotel_qty = []

for rating in ratings:
  rat = rating.get_text(strip=True)
  # if rat[0:2] == '10':
  #   rat1 = rat[0:2]
  # else:
  #   rat1 = rat[0:3]
  rat1 = rat[0:3]
  rat2 = rat.split("(")[-1].replace(")", "")
  rat3 = rat[3:].split("(")[0]
  rating_hotel_num.append(rat1)
  rating_hotel_text.append(rat3)
  rating_hotel_qty.append(rat2)

  
print(rating_hotel_num)
print(rating_hotel_qty)
print(rating_hotel_text)
```

```python
# discount price
discount_prices = soup.find_all("div", class_="diskonPrice")

discount_price_hotel = []

for discount_price in discount_prices:
  dis_prc = discount_price.get_text(strip=True)
  dis_prc = dis_prc.replace("Rp ", "")
  discount_price_hotel.append(dis_prc)

print(discount_price_hotel)
```

```python
# reward
rewards = soup.find_all("span", class_="pepereward-text")

reward_point_hotel = []

for reward in rewards:
  rwd = reward.get_text(strip=True)
  rwd = rwd.split("Reward ")[1].replace(" Poin", "")
  reward_point_hotel.append(rwd)

print(reward_point_hotel)
```
